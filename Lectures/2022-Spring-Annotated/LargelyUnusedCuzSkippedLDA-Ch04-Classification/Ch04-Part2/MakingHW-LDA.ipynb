## Making a homework problem to go with section 4.4.1

```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
mu1 = 0
mu2 = 2
sigma = 1

t = np.linspace(-6,6,100)
plt.plot(t,stats.norm.pdf(t,mu1,sigma))
plt.plot(t,stats.norm.pdf(t,mu2,sigma))
```

```python
np.random.seed(48824)
n1 = 30
n2 = 20
X1 = np.random.normal(mu1,sigma**2, n1)
X2 = np.random.normal(mu2,sigma**2,n2)
```

```python
XY1 = np.concatenate([X1.reshape(1,-1),np.ones(n1).reshape(1,-1)],axis = 0).T
XY2 = np.concatenate([X2.reshape(1,-1),np.ones(n2).reshape(1,-1)],axis = 0).T
XY2[:,1] = 2

alldata = np.concatenate([XY1,XY2], axis = 0)
# alldata
```

```python
# np.savetxt('LDA_example.csv',alldata,fmt='%1.3f', delimiter = ' , ')
```

```python

plt.hist(X1)
plt.hist(X2)
```

```python
mu1hat = np.average(X1)
mu1hat
```

```python
mu2hat = np.average(X2)
mu2hat
```

```python
sigma2hat = (np.sum((X1-mu1hat)**2) + np.sum((X2-mu2hat)**2))/(n1+n2-2)
sigma2hat
```

```python
pi1hat = n1/(n1+n2)
pi2hat = n2/(n1+n2)
```

```python
delta1hat = t*mu1hat/sigma2hat - mu1hat**2/(2*sigma2hat) + np.log(pi1hat)
delta2hat = t*mu2hat/sigma2hat - mu2hat**2/(2*sigma2hat) + np.log(pi2hat)
plt.plot(t,delta1hat)
plt.plot(t,delta2hat)

# plt.plot(t,stats.norm.pdf(t,mu1,sigma))
# plt.plot(t,stats.norm.pdf(t,mu2,sigma))
```

```python
np.where(np.abs(delta1hat-delta2hat)<0.1)
```

```python
# Location for the approximate decision boundary
t[58]
```

```python
delta1 = t*mu1/sigma - mu1**2/(2*sigma) #+ np.log(pi1hat)
delta2 = t*mu2/sigma - mu2**2/(2*sigma) #+ np.log(pi2hat)
plt.plot(t,delta1)
plt.plot(t,delta2)
```

```python
np.where(np.abs(delta1-delta2)<0.2)
```

```python
# Location for the true decision boundary
t[57]
```

```python
plt.plot(t,stats.norm.pdf(t,mu1,sigma))
plt.plot(t,stats.norm.pdf(t,mu2,sigma))

plt.plot([t[58],t[58]],[0,1], label = 'approximate')
plt.plot([t[57],t[57]],[0,1], label = 'truth')

plt.legend()
```
