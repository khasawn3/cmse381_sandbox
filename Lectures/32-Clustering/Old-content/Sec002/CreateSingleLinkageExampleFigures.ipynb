```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy.spatial import distance_matrix
from scipy.spatial.distance import pdist
from scipy.cluster import hierarchy

import pylab as pl
```

```python
np.random.seed(1)
X = np.random.random((10,2))
plt.scatter(X[:,0],X[:,1])
```

```python
D = distance_matrix(X,X)
```

```python
plt.matshow(D)
plt.colorbar()

plt.savefig('ExFigs/SingleLink-Distance.pdf',bbox_inches = 'tight')
```

```python
plt.scatter(X[:,0],X[:,1])

r = 0.3

for i in range(10):
    for j in range(10):
        if D[i,j] <= r:
            p = X[(i,j),:]
            plt.plot(p[:,0],p[:,1],c = 'black')
```

```python

for r in [0,0.05,0.1,0.15,0.2,0.25]:

    plt.scatter(X[:,0],X[:,1])
    eps = 0.01
    for i in range(10):
        pl.text(X[i,0]+eps,X[i,1]-3*eps,str(i),color="black", fontsize=14)
    pl.margins(0.1)



    for i in range(10):
        for j in range(10):
            if D[i,j] <= r:
                p = X[(i,j),:]
                plt.plot(p[:,0],p[:,1],c = 'black')
    
    plt.title('r ='+str(r))
    plt.savefig('ExFigs/SingleLink-0p'+str(int(100*r))+'.pdf',bbox_inches = 'tight')
    plt.clf()
```

```python


Z = hierarchy.linkage(pdist(X), 'single')

plt.figure()

dn = hierarchy.dendrogram(Z)

plt.savefig('ExFigs/SingleLink-Dendrogram.pdf',bbox_inches = 'tight')
```
