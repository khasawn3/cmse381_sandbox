```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
n = 50

def f(x):
    return 3*x -2 


X = np.random.random(n)*5+1
X.sort()
X

Y = [f(x) for x in X] + np.random.normal(0, 5, size = n)

plt.scatter(X,Y)
plt.plot([1,6], [f(1),f(6)], color = 'red')

plt.savefig('ExampleLinear50pts.png', bbox_inches = 'tight', dpi = 100)
```
